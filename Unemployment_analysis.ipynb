# Visualizing Data

```python
ax=sns.countplot(x='Region', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

```python
plt.figure(figsize=(12,10))
plt.title('Unemployment In India State Wise')
sns.histplot(x=' Estimated Unemployment Rate (%)', hue="Region", data=df)
plt.show()
```

```python
#Highest Unemployment rate by state
group = df.groupby(['Region'])['EstimatedUnemploymentRate(%)'].max()
group
```

```python
group = df.groupby(["Region"])["EstimatedUnemploymentRate(%)"].aggregate("max").reset_index()
plt.figure(figsize=(12,8))
sns.pointplot(x=group['Region'], y=group['EstimatedUnemploymentRate(%)'])
plt.ylabel('Max rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Highest Rates of Unemployment by States", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```
